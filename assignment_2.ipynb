### Question 1:

> **You have a coin that you think is biased, you flip it 5 times and get the sequence HT HT H. What is the maximum likelihood estimate for the probability of getting heads?**

**Answer:**

Given the sequence of coin flips HT HT H, we have a total of \( n = 5 \) trials and \( k = 3 \) successes (heads). The likelihood function \( L \) under a Bernoulli distribution is:

$$
L(p) = p^k(1-p)^{n-k}
$$

Plugging the values, we get:

$$
L(p) = p^3(1-p)^2
$$

To find the MLE, we need to maximize this function. We can do this by taking the derivative of \( L(p) \) with respect to \( p \), setting it equal to zero, and solving for \( p \):

$$
\frac{d}{dp} L(p) = \frac{d}{dp} \left(p^3(1-p)^2\right) = 0
$$

Now, solve for \( p \):

$$
3p^2(1-p)^2 + p^3(-2)(1-p) = 0
$$

$$
3p^2(1-p)^2 - 2p^3(1-p) = 0
$$

The critical points for this equation are \( p = 0, \, p = 1, p = $\frac{2}{3}$ \).

To determine the nature of these critical points, use the second derivative test. The second derivative of \( L(p) \) is:

$$
\frac{d^2}{dp^2} L(p) = \frac{d^2}{dp^2} \left(p^3(1-p)^2\right)
$$

Evaluating this second derivative at each of the critical points \( p = 0, \, p = 1, p = $\frac{2}{3}$ \) will help to identify which of these points maximizes the likelihood function.

By doing this, you'll find that at \( p = $\frac{2}{3} $\), the second derivative is negative, indicating a local maximum. Therefore, the Maximum Likelihood Estimate (MLE) for the probability of getting heads is \( p = $\frac{2}{3}$ \).


### Question 2:

> **You know that candy prices are normally distributed with mean µ and standard deviation σ. You have three candy pricing 2, 3, 5 lira. What is the maximum likelihood for µ?**

**Answer:**

Given the candy prices of 2, 3, and 5 lira, and knowing that the candy prices are normally distributed with mean $\mu$ and standard deviation $\sigma$, we want to find the Maximum Likelihood Estimate (MLE) for $\mu$.

The probability density function (pdf) of a normal distribution is given by:

$$
f(x | \mu, \sigma) = \frac{1}{{\sigma\sqrt{2\pi}}} e^{-\frac{1}{2} \left(\frac{x - \mu}{\sigma}\right)^2}
$$

The likelihood function $\mu$, $\sigma$ for the observed data is the product of the pdf evaluated at each data point:

$$
L(\mu, \sigma) = f(2 | \mu, \sigma) \cdot f(3 | \mu, \sigma) \cdot f(5 | \mu, \sigma)
$$

Taking the natural logarithm of the likelihood function gives the log-likelihood function, which is easier to work with:

$$
\ln(L(\mu, \sigma)) = -\frac{3}{2} \ln(2\pi) - 3 \ln(\sigma) - \frac{1}{2\sigma^2} \left((2 - \mu)^2 + (3 - \mu)^2 + (5 - \mu)^2\right)
$$

We are interested in finding the value of$\mu$ that maximizes the log-likelihood function. To do this, we take the derivative of the log-likelihood function with respect to $\mu$, set it equal to zero, and solve for $\mu$:

$$
\frac{d}{d\mu} \ln(L(\mu, \sigma)) = 0
$$

Solving the above equation yields the Maximum Likelihood Estimate (MLE) for $\mu$. In this case, we can simplify the equation to get:

$$
0 = (2 - \mu) + (3 - \mu) + (5 - \mu)
$$

Now, solve for  $\mu$ :

$$
0 = 10 - 3\mu
$$

$$
3\mu = 10
$$

$$
\mu = \frac{10}{3}
$$

So, the Maximum Likelihood Estimate (MLE) for $\mu$ is $\frac{10}{3}$ or approximately $ 3.33 $ lira.

### Question 4:

> **Suppose that X1, ..., Xn form a random sample from a uniform distribution on the interval (0, θ), where of the parameter θ > 0 but is unknown. Please find MLE of θ.**

**Answer:**

### Maximum Likelihood Estimate (MLE) for $\theta$ in a Uniform Distribution

Given a random sample $X_1, X_2, ..., X_n$ from a uniform distribution on the interval $(0, \theta)$, we aim to find the Maximum Likelihood Estimate (MLE) for $\theta$.

The probability density function (pdf) of a uniform distribution on this interval is given by:

$$
f(x | \theta) = \frac{1}{\theta} \quad \text{for } 0 < x < \theta
$$

The likelihood function $L(\theta)$ for the given sample is the product of the pdf evaluated at each data point:

$$
L(\theta) = \left(\frac{1}{\theta}\right)^n
$$

However, this expression holds only if all $x$ values are within the interval $(0, \theta)$. Thus, the likelihood function is actually zero for any $\theta$ less than the maximum observed value $x_{(n)}$ (the $n$th order statistic). So, we should express the likelihood function as:

$$
L(\theta) = \left(\frac{1}{\theta}\right)^n \quad \text{for } \theta > x_{(n)}
$$

Taking the natural logarithm of the likelihood function gives the log-likelihood function:

$$
\ln(L(\theta)) = -n \ln(\theta)
$$

Now, to find the MLE for $\theta$, we need to maximize the log-likelihood function $\ln(L(\theta))$. The derivative of $\ln(L(\theta))$ with respect to $\theta$ is:

$$
\frac{d}{d\theta} \ln(L(\theta)) = -\frac{n}{\theta}
$$

This derivative does not set to zero as there is no maximum point for $\theta$ in this function. Instead, the function is monotonically decreasing, meaning that $L(\theta)$ is maximized when $\theta$ is minimized.

Given the condition that all $x$ values must be within the interval $(0, \theta)$, the smallest possible value of $\theta$ that satisfies this condition is $\theta = x_{(n)}$, the maximum observed value in the sample.

Therefore, the Maximum Likelihood Estimate (MLE) for $\theta$ is $\theta_{MLE} = x_{(n)}$.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
classes = np.load('classes.npy')
train_data = np.load('train_data.npy')
train_labels = np.load('train_labels.npy')
train_masks = np.load('train_mask.npy')
```

```python
# Exclude the areas from analysis using the mask. We'll set the value to NaN where the mask is 0.
train_data_masked = np.where(train_masks[:, :, np.newaxis] == 0, np.nan, train_data)

# Split the data into 6 spectral bands
# Since our data is in the shape (height, width, channels), we can split it along the last axis
bands = np.split(train_data_masked, train_data_masked.shape[-1], axis=-1)

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Define color maps for each band. For infrared, we use a 'hot' colormap to represent intensity.
color_maps = ['Reds', 'Greens', 'Blues', 'hot', 'hot', 'hot']

for i, ax in enumerate(axs.flat):
    band = bands[i].reshape(train_data.shape[0], train_data.shape[1]) # reshape the band to (9393, 5642)
    
    # Use the 'gray' color map for the VNIR bands and a 'hot' color map for infrared bands
    cmap = color_maps[i] if i < 3 else 'hot'
    ax.imshow(band, cmap=cmap)
    ax.set_title(f'Band {i+1}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
class GaussianNB:
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.parameters = {}
        
        for c in self.classes:
            # Calculate mean, variance and prior for each class
            X_c = X[y == c]
            self.parameters[c] = {
                "mean": X_c.mean(axis=0),
                "var": X_c.var(axis=0),
                "prior": X_c.shape[0] / X.shape[0]
            }

    def _calculate_log_probability(self, X, mean, var):
        # Using the Gaussian log probability distribution function
        exponent = -0.5 * np.sum(np.log(2. * np.pi * var))
        exponent -= 0.5 * np.sum(((X - mean) ** 2) / (2 * var), 1)
        return exponent

    def _predict_log_proba(self, X):
        # Calculate the log probability of each class
        log_probs = np.zeros((X.shape[0], len(self.classes)))
        
        for c, params in self.parameters.items():
            log_prior = np.log(params["prior"])
            log_probability = self._calculate_log_probability(X, params["mean"], params["var"])
            log_probs[:, c] = log_probability + log_prior
            
        return log_probs

    def predict(self, X):
        # Predict the class with the highest log probability
        log_probs = self._predict_log_proba(X)
        return self.classes[np.argmax(log_probs, axis=1)]
```

```python
def accuracy_score(labels_test, predictions):
    return ((labels_test == predictions).sum() / len(labels_test)) * 100 # return the accuracy score in percentage
```

```python
train_data = train_data[train_masks != 0] # Remove the masked areas from the training data
train_labels = train_labels[train_masks != 0] # Remove the masked areas from the training labels
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    train_data,  # Features
    train_labels,  # Labels
    test_size=0.3,  # 30% for testing
    random_state=42  # Seed for reproducibility
)

# Function to train and evaluate a Naive Bayes model
def train_evaluate_naive_bayes(features_train, labels_train, features_test, labels_test):
    model = GaussianNB()
    model.fit(features_train, labels_train)
    predictions = model.predict(features_test)
    return accuracy_score(labels_test, predictions)

# 1. Classify objects with Naive Bayes for each spectral band
acc_per_band = []
for i in range(train_data.shape[-1]):
    band_i = X_train[:, i:i+1]  # Selecting the ith band
    test_band_i = X_test[:, i:i+1]
    accuracy = train_evaluate_naive_bayes(band_i, y_train, test_band_i, y_test)
    acc_per_band.append(accuracy)
    print(f'Accuracy for band {i+1}: {accuracy}')

# 2. Generate the RGB image and classify objects with Naive Bayes
rgb_bands = X_train[:, :3]  # First three bands for RGB
test_rgb_bands = X_test[:, :3]
accuracy_rgb = train_evaluate_naive_bayes(rgb_bands, y_train, test_rgb_bands, y_test)
print(f'Accuracy for RGB bands: {accuracy_rgb}')

# 3. Generate the Infrared image and classify objects with Naive Bayes
ir_bands = X_train[:, 3:]  # Last three bands for Infrared
test_ir_bands = X_test[:, 3:]
accuracy_ir = train_evaluate_naive_bayes(ir_bands, y_train, test_ir_bands, y_test)
print(f'Accuracy for Infrared bands: {accuracy_ir}')

# 4. Use all six spectra to classify objects with Naive Bayes
accuracy_all_bands = train_evaluate_naive_bayes(X_train, y_train, X_test, y_test)
print(f'Accuracy for all bands: {accuracy_all_bands}')
```
